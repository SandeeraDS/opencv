```python
#all the 6 method comparison in a list
#Note how we are using string, later on we'll use the eval() function to covert  to function

methods = ['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    
    #create a copy
    full_copy = full.copy()
    
    method = eval(m)
    
    #template matching
    res = cv2.matchTemplate(full_copy,face,method)
    
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]: #use minimum to find out a match
        top_left=min_loc #(x,y)
    else:
        top_left=max_loc
        
    height,width,channels = face.shape
    
    bottom_right = (top_left[0]+width,top_left[1]+height)
    
    cv2.rectangle(full_copy,top_left,bottom_right,(255,0,0),10)
    
    #plot  and show the images
    
    plt.subplot(121) # 1row 2 col and this is the 1st col
    plt.imshow(res)
    plt.title('HEATMAP OF  TEMPLATE MATCHING')
    
    plt.subplot(122) # 2nd col
    plt.imshow(full_copy)
    plt.title('DETECTION OF TEMPLATE')
    #TITLE WITH THE METHOD USED
    plt.suptitle(m)
    
    plt.show() #to avoiding overiding subplots
    print('\n')    
    print('\n')
```
